## 데이터 프레임 병합
- merge() 함수는 sql의 join 명령과 비슷한 방식으로 어떤 기준에 의해 두 데이터프레임을 병합하는 개념
- 기준이 되는 열이나 인덱스를 키라고 부름
- 키가 되는 열이나 인덱스는 반드시 양쪽 데이터프레임에 모두 존재 함

```python
import pandas as pd

# filepath1 = 
# contents/dataset/stock price.xlsx
# filepath2 = './dataset/stock valuation.xlsx'
# 주식 데이터를 가져와서 데이터 프레임 만들기
df1 = pd.read_excel('stock price.xlsx')
df2 = pd.read_excel('stock valuation.xlsx')

display(df1)
print('\n')
display(df2)
print('\n')
```

```python
# 데이터프레임 합치기 - 교집합
# on = 'None', how = 'inner'옵션이 기본값으로 적용
# on = 'None'옵션은 두 데이터프레임에 공통으로 속하는 모든 열을 기준(키)로 병합한다는

merge_inner = pd.merge(df1, df2)
display(merge_inner)
```

#### Q. df1, df2의 합집합을 구한후 출력하세요.

```python
# 데이터프레임 합치기 - 왼쪽 데이터프레임 기준, 키 값 분리
# how='left' : 왼쪽 df의 키 열에 속하는 데이터 값 기준으로 병합
# left_on, right_on : 죄우 df에 각각 다르게 키를 지정, id_x, id_y 구분하여 표시

merge_left = pd.merge(df1, df2, how='left', left_on='stock_name', right_on='name')
display(merge_left)

merge_right = pd.merge(df1, df2, how='right', left_on='stock_name', right_on='name')
display(merge_right)
```

```python
# df1에서 불린 인덱싱과 결합하여 원하는 데이터 찾기

price = df1[df1['price'] < 50000]
display(price.head())
```

#### Q. df2에서 per가 30보다 크고 pbr이 0.8보다 큰 데이터를 모두 출력하세요.

```python
pcon = df2[(df2.per > 30) & (df2.pbr > 0.8)]
display(pcon)
```

## 데이터프레임 결합
- 판다스 join() 메소드는 merge() 함수를 기반으로 만들어졌기 때문에 기본 작동 방식이 서로 비슷
- join() 메소드는 두 데이터 프레임의 행 인덱스를 기준으로 결합하는 점에서 merge와 차이가 있으나 on=keys 옵션 설정하면 열을 기준으로 결합

```python
df1 = pd.read_excel('stock price.xlsx', index_col = 'id')
df2 = pd.read_excel('stock valuation.xlsx', index_col='id')

display(df1)
print('\n')
display(df2)
print('\n')
```

```python
# 데이터프레임 결합(join)
# df1의 행 인덱스를 기준으로 결합하는 how='left' 옵션이 기본 적용
df3 = df1.join(df2)
display(df3)
```
