## 모듈
- 특정 기능들(함수, 변수, 클래스)이 구현되어 있는 파이썬 파일

## 패키지
 - 특정 기능과 관련된 여러 모듈들을 하나의 상위 폴더에 넣어놓은 것을 패키지라 한다.
 - 패키지 안에 여러가지 폴더가 더 존재할 수 있다.
 - 패키지를 표현해주기 위해 __init__.py 가 존재해야 한다.
 - 파이썬 3.3부터는 __init__.py 가 없어도 패키지로 인식하나, 하지만 하위 버전 호환을 위해 init.py파일 생성하는 것이 안전
 
## 라이브러리
 - 여러 모듈과 패키지를 묶어 라이브러리라고 한다.
 - 파이썬을 설치할 때 기본적으로 설치되는 라이브러리를 표준 라이브러리(Python standard Library)
 - 파이썬 공식이 아닌 외부(3rd party)에서 개발한 모듈과 패키지를 묶어 외부 라이브러리

## 빅데이터 분석에서 판다스, 넘파이를 알아야 하는 이유
- 빅데이터 분석을 쉽게 해주는 도구로 파이썬, 판다스를 사용하는 최근 동향
- 판다스는 파이썬으로 만들어진 프레임 워크
- 파이썬 문법으로 사용할 수 있는 도구의 집합체

- 넘파이는 C언어로 구현된 파이썬 바인딩 수학 패키지(프레임워크)
- 데이터를 다루면서 벡터연산(한번에 여러 데이터를 동시에 넘겨서 동시처리)이 매우 유용
- For문으로 처리할 것을 numpy를 사용하면 데이터를 벡터연산으로 처리하여 성능 향상

- 빅데이터 관련 파이썬이 대세가 되는 이유는 이런 라이브러리 때문임

## 판다스
- 판다스는 추상화 된 도구이며 간단한 함수로 복잡한 기능을 수행하는 반면
  이런 추상화는 내부 자료형의 크기를 크게 만들어 많은 메모리를 필요로 함
- 판다스 사용을 위해서는 라이브러리 설치 후 import 필요.
  간결성을 위해 pd라는 축약된 이름 사용 : import pandas as pd
- 판다스의 데이터 구조는 Series와 Dataframe
- 파이썬의 리스트, 넘파이 등의 내부 데이터는 물론 CSV 등의 파일을 쉽게 데이터 프레임으로 변경해 데이터 가공분석을 편리하게 수행
- 인덱스는 개별 데이터를 고유하게 식별하는 Key값으로 Series, DataFrame 모두 인덱스를 Key값으로 가지고 있음.

## 시리즈, 데이터프레임
- 시리즈는 데이터가 순차적으로 나열됨 1차원 배열의 형태

- 인덱스(index)는 데이터 값(value)과 일대일 대응이 된다. 이런 관점에서 키(k)와 값(v)이 '{k:v}' 형태로
짝을 이루는 파이썬 딕셔너리(dictionary)와 비슷한 구조

- 시리즈의 인덱스는 데이터 값의 위치를 나타내는 이름표(데이터 주소) 역할
예를들어, 'index 2'라는 주소를 알고 있다면 'Data 2'라는 원소 데이터 값에 바로 접근 가능.

- 데이터 프레임은 2차원 배열 행과 열로 만들어지는 2차원 배열 구조는 마이크로소프트 엑셀과 관계형 데이터베이스(RDBMS)등 컴퓨터 관련 다양한 분야에서 사용
- 판다스의 데이터프레임 자료구조는 대표적인 통계 패키지인 R의 데이터프레임에서 유래
- 여러 개의 시리즈들이 한데 모여서 데이터프레임을 이루는 구조로 데이터프레임의 열은 각각 시리즈 객체
 
- 시리즈를 열벡터라고 하면, 데이터프레임은 여러 개의 열벡터들이 같은 행 인덱스를 기준으로 줄지어 결합된 2차원 벡터 또는 행렬
