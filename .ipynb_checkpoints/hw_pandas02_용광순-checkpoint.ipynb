### Q. 연속된 정수 1 ~ 24 와 25~50  으로 각각 구성된 (8,3) 행렬 데이터프레임 df_a와 df_b를 생성 출력하세요.

```python
import numpy as np
import pandas as pd

list1 = np.arange(start=1, stop=25)
list2 = np.arange(start=25, stop=49)

df_a = pd.DataFrame(list1.reshape(8,3), columns=["x","y","z"])
df_b = pd.DataFrame(list2.reshape(8,3), columns=["x","y","z"])

display(df_a)
display(df_b)
```

### Q. df_a, df_b의 사칙연산 결과를 데이터프레임으로 합쳐서 출력하세요.

```python
add = df_a + df_b
sub = df_a - df_b
mul = df_a * df_b
div = round((df_a / df_b), 2)
columns = ["x", "y", "z"]

result = pd.concat([add, sub, mul, div],keys= ["덧셈", "뺄셈", "곱셈", "나눗셈"], axis = 1)
display(result)
```

### Q. df_a,df_b 데이터 프레임으로 아래 사항을 수행하여 df_na와 df_nb를 생성하세요.
- 4행에서 끝행까지 선택 추출한 후 df_an, df_bn으로 각각 저장
- df_an, df_bn의 인덱스를 각각 리셋한 후 df_na, df_nb로 저장
- df_na의 세번째 행 x 위치에 np.nan 적용
- df_nb의 네번째 행, z 위치에 np.nan 적용
- df_na, df_nb를 출력, NaN 값 확인

```python
df_an = df_a[4:]
df_bn = df_b[4:]
# print(df_an)
df_na = df_an.reset_index(drop=True, inplace=False)
df_nb = df_bn.reset_index(drop=True, inplace=False)
display(df_na)

df_na.iloc[2,0] = np.nan
display(df_na)

df_nb.iloc[3,2] = np.nan
display(df_nb)
```

### Q. df_na와 df_nb의 NaN 값을 0으로 대체하고 사칙연산을 수행하세요.

```python
add = df_na.add(df_nb, fill_value=0)
sub = df_na.sub(df_nb, fill_value=0)
mul = df_na.mul(df_nb, fill_value=0)
div = df_na.div(df_nb, fill_value=0)

result = pd.concat([add, sub, mul, div], keys= ["덧셈", "뺄셈", "곱셈", "나눗셈"], axis = 1)
display(result)
```
