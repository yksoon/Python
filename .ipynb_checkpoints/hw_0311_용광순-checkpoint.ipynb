#### 1. 다음은 학생 성적 관리 프로그램으로 리스트와 딕셔너리로 표현하고 for문을 사용해서 학생별 총점과 평균을 구했다. 딕셔너리로 객체를 각각 만드니 번거롭고 실수가 발생될 가능성이 높았다. Student 클래스를 만들고 각 학생 객체로 구성되는 학생 리스트를 작성하여 for문을 이용하여 출력하는 방식으로 프로그램을 개선 하세요.

```python
# 학생 리스트를 선언합니다.
students = [
    { "name": "조인성", "korean": 87, "math": 98, "english": 88, "science": 95 },
    { "name": "연하진", "korean": 92, "math": 98, "english": 96, "science": 98 },
    { "name": "구지연", "korean": 76, "math": 96, "english": 94, "science": 90 },
    { "name": "나선주", "korean": 98, "math": 92, "english": 96, "science": 92 },
    { "name": "윤아린", "korean": 95, "math": 98, "english": 98, "science": 98 },
    { "name": "윤명월", "korean": 64, "math": 88, "english": 92, "science": 92 }
]

# 학생을 한 명씩 반복합니다.
print("이름", "총점", "평균", sep="\t")
for student in students:
    # 점수의 총합과 평균을 구합니다.
    score_sum = student["korean"] + student["math"] +\
        student["english"] + student["science"]
    score_average = score_sum / 4
    # 출력합니다.
    print(student["name"], score_sum, score_average, sep="\t")
```

```python
# ****풀이****

class Student:
    def __init__(self, name, korean, math, english, science):
        self.name = name
        self.korean = korean
        self.math = math
        self.english = english
        self.science = science
    
    def get_sum(self):
        return self.korean + self.math + self.english + self.science
    
    def get_average(self):
        return self.get_sum() / 4
    
    def to_string(self):
        return "{}\t{}\t{}".format(self.name, self.get_sum(), self.get_average())
    
students = [
    Student("조인성", 87, 98, 88, 95),
    Student("연하진", 92, 98, 96, 98),
    Student("구지연", 76, 96, 94, 90),
    Student("나선주", 98, 92, 96, 92),
    Student("윤아린", 95, 98, 98, 98),
    Student("윤명월", 64, 88, 92, 92),
]

print("이름", "총점", "평균", sep="\t")
for student in students:
    print(student.to_string())
```

#### 2. 원의 둘레와 넓이를 구하는 메소드를 포함하고 있는 Circle 클래스를 활용해서 반지름이 10인 원의 둘레와 넓이를 구하세요. 

```python
class Circle:
    
    pi = 3.141592
    
    def __init__(self, r):
        self.r2 = r
    
    def circ(self):
        self.l = 2 * self.pi * self.r2
        return self.l
    
    def area(self):
        self.s = self.pi * (self.r2 ** 2)
        return self.s

circle = Circle(10)
print("반지름이 %d인 원의 둘레 : %f" % (10, circle.circ()))
print("반지름이 %d인 원의 넓이 : %f" % (10, circle.area()))
```

```python
# ****풀이****

# 모듈을 가져옵니다.
import math

# 클래스를 선언합니다.
class Circle:
    def __init__(self, radius):
        self.radius = radius
    def get_circumference(self):
        return 2*math.pi *self.radius
    def get_area(self):
        return math.pi * (self.radius ** 2)

# 원의 둘레와 넓이를 구합니다.
circle = Circle(10)
print("원의 둘레 : ", circle.get_circumference())
print("원의 넓이 : ", circle.get_area())
```

#### 4. 3번에서 작성한 1000명의 키와 몸무게로 BMI(비만도)를 계산해서 저체중, 과체중, 정상체중으로 분류해서 출력하세요.
BMI = int(weight) / ((int(height) / 100) **2)

```python
# ****풀이****

with open("info.txt", "r") as file:
    for line in file:
        # 변수를 선언합니다.
        (name, weight, height) = line.strip().split(", ")
        
        # 데이터가 문제 없는지 확인합니다: 문제가 있으면 지나감
        if (not name) or (not weight) or (not height):
            continue
        
        # 결과를 계산합니다.
        bmi = int(weight) / ((int(height) / 100) ** 2)
        result = ""
        if 25 <= bmi:
            result = "과체중"
        elif 18.5 <= bmi:
            result = "정상체중"
        else:
            result = "저체중"
        
        #출력합니다.
        print("\n".join([
            "이름: {}",
            "몸무게: {}",
            "키: {}",
            "BMI: {}",
            "결과: {}"
        ]).format(name, weight, height, bmi, result))
        print()
```

#### 5. 재귀함수를 이용해 다음 팩토리얼을 출력하세요.
1!: \
2!: \
3!: \
4!: \
5!: 

```python
def factorial(n):
    if n > 0:
        return n * factorial(n-1)
    else:
        return 1

for i in range(1, 6):
    print("%d! : %d" % (i, factorial(i)))
```

```python
# ****풀이****

# 함수를 선언합니다.
def factorial(n):
    # n이 0이라면 1을 리턴
    if n == 0:
        return 1
    # n이 0이 아니라면 n * (n-1)!을 리턴
    else:
        return n * factorial(n-1)

print("1! :", factorial(1))
print("2! :", factorial(2))
print("3! :", factorial(3))
print("4! :", factorial(4))
print("5! :", factorial(5))
```
